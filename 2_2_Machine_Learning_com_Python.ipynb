#Scikit-Learn

```python
import pandas as pd
```

```python
! pip install scikit-learn
```

```python
df_diabetes = pd.read_csv("/content/diabetes.csv")
```

```python
df_diabetes.head()
```

```python
from sklearn.model_selection import train_test_split
#parte da biblioteca do scikit learn que nos ajuda a separar os dados em módulos
#a seguir separamos os módulos em x e y
```

```python
x = df_diabetes.drop(["Class variable"], axis=1)
```

```python
x.head()
```

```python
y = df_diabetes["Class variable"]
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
#separamos 30% de amostra para teste
```

```python
len(x_train)
```

```python
x_train.head()
```

```python
len(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
#pelo que entendi aqui dizemos para o algoritmo que ele vai se basear em 3 entradas para encontrar similaridades
```

```python
knn.fit(x_train, y_train) #treinando a base
```

```python
accuracy = knn.score(x_test, y_test) #testando a acuracia do algoritmo
```

```python
accuracy
```
